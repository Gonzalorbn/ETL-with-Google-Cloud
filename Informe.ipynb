# **ETL CON GOOGLE CLOUD**

## **PROPÓSITO DEL PROYECTO**
Este proyecto se centra en la creación de un sistema ETL utilizando Google Cloud para aprovechar las ventajas que poseen sus servicios en el tema de almacenamiento orquestación y análisis. Comienza con la extracción de información en la API de Mercado Libre sobre la categoría “consolas”, transformamos los datos extraídos dejando solo la información relevante, finalmente se almacena en Cloud Storage y carga en una tabla de BigQuery para análisis.

## **DETALLES DEL PROYECTO**

### **TECNOLOGÍAS UTILIZADAS**


*   Apache Airflow: Utilizado como el motor de ejecución de tareas y programación de trabajos para automatizar flujos de trabajo.

*   Visual Studio Code (VSCode): Utilizado como entorno de desarrollo para escribir, depurar y administrar código.
*   Postman: Herramienta utilizada para probar y depurar APIs.


*   Cloud Storage: Se empleó Cloud Storage para almacenar los datos extraídos de la API de Mercado Libre antes de cargarlos en BigQuery.


*   BigQuery: BigQuery se utilizó como el almacén de datos para realizar análisis eficientes y consultas SQL en los datos almacenados.
*   Google Cloud Composer: Google Cloud Composer se utilizó como la plataforma de orquestación de tareas para programar y ejecutar el flujo de trabajo ETL.





### **DESCRIPCIÓN DEL FLUJO DE TRABAJO**
### **Extracción de Datos:**


> En el archivo "consulta_api_dag.py", se define una tarea de inicio y otra llamada “Subir a Cloud Storage”, la misma se encarga de llamar a la función “get_most_relevant_items_for_category
” que devuelve una lista de diccionarios “relevant_data” con la información relevante de la API

### **Transformación de Datos:**


> Dentro de la misma función “subir_a_cloud_storage” se lleva a cabo una corrección de valores “null” para que no genere conflictos al momento de subir a BigQuery
Se prepara una lista de líneas que contienen los datos en un formato de archivo TSV (Tab-Separated Values).
### **Almacenamiento en Cloud Storage:**


> La función "upload_to_cloud_storage" se encarga de cargar los datos transformados en Cloud Storage de Google Cloud.
Se utiliza el cliente de Cloud Storage para crear un objeto Blob y subir los datos a la nube.

### **Carga en BigQuery:**


> Finalmente tenemos la tarea llamada “Cargar a BigQuery” cuya función es cargar los datos que se guardaron en un archivo “File.tsv” a una tabla en BigQuery
